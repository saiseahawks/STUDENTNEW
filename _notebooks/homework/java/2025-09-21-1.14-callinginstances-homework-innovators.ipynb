## Popcorn Hack #2

- Add on to the previous counter class and add three more methods for subtraction, multiplication, and division.

- Call all five methods outside of the class.

````java
// Work by Sai Talisetty

public class Counter {
    private int count;

    // Constructor
    public Counter(int start) {
        count = start;
    }

    // Add method
    public void add(int value) {
        count += value;
    }

    // Subtract method
    public void subtract(int value) {
        count -= value;
    }

    // Multiply method
    public void multiply(int value) {
        count *= value;
    }

    // Divide method (with a simple check to avoid dividing by zero)
    public void divide(int value) {
        if (value != 0) {
            count /= value;
        } else {
            System.out.println("Cannot divide by zero!");
        }
    }

    // Reset method (just for fun)
    public void reset() {
        count = 0;
    }

    // Display the current count
    public void printCount() {
        System.out.println("Current count: " + count);
    }

    // Main method to test
    public static void main(String[] args) {
        // Create a Counter object
        Counter myCounter = new Counter(10);

        // Call all five methods
        myCounter.add(5);        // 10 + 5 = 15
        myCounter.subtract(3);   // 15 - 3 = 12
        myCounter.multiply(2);   // 12 * 2 = 24
        myCounter.divide(4);     // 24 / 4 = 6
        myCounter.reset();       // set to 0

        // Print the final result
        myCounter.printCount();
    }
}
Counter.main(null);
````
